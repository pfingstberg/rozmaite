```python
def twopict(pict1,pict2):
    fig = plt.figure()
    fig.suptitle(".twopict.")
    ax1 = plt.subplot(1, 2, 1)
    ax2 = plt.subplot(1, 2, 2)
    ax1.axes.xaxis.set_ticks([])
    ax1.axes.yaxis.set_ticks([])
    ax2.axes.xaxis.set_ticks([])
    ax2.axes.yaxis.set_ticks([])
    ax1.set_title(".left.")
    ax2.set_title(".right.")
    image1 = mpimg.imread(pict1)
    image2 = mpimg.imread(pict2)
    img1 = ax1.imshow(image1)
    img2 = ax2.imshow(image2)
    return
def twocont(pict1,pict2,ax1,ax2):
    image1 = mpimg.imread(pict1)
    image2 = mpimg.imread(pict2)
    ax1.imshow(image1)
    ax2.imshow(image2)
    return
```
